'''
porphyrin angle distance analysis.

'''

```python
import os
import prody as pr

hem_database = '/Users/lonelu/DesignData/ligands_metal/porphyrin/pdbs/M1_AAMetal_HIS_reps/'

pdbs = [pr.parsePDB(hem_database + x) for x in os.listdir(hem_database)]

#angles between NE2-FE-NA
angles = []
for pdb in pdbs:
    print(pdb.getTitle())
    a = pr.calcAngle(pdb.select('name NE2'), pdb.select('name FE NI ZN CU')[0], pdb.select('name NA'))
    angles.append(a[0])


dihedrals = []
for pdb in pdbs:
    print(pdb.getTitle())
    a = pr.calcDihedral(pdb.select('name CE1'), pdb.select('name NE2'), pdb.select('name FE NI ZN CU')[0], pdb.select('name NA'))
    dihedrals.append(a[0])

dihedrals_norm = [(d+360)%90 for d in dihedrals]


```

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(angles, kde=True, bins=50)
plt.show()
```

```python
sns.histplot(dihedrals_norm, kde=True, bins=50)
plt.show()
```
